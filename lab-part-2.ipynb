```python
# cosine similarity of words in each time step

experimentData2 = np.asarray(data['E'])

withinEachCosineSimilarities = []
for i in range(experimentData.shape[1]):
    withinEachCosineSimilarities.append(cosine_similarity(experimentData2[:, i, :]))
```

```python
def getRelatedWords(wordIndex):
    similarities = []
    for i in range(10):
        inyearsimilarity = []
        for j, word in enumerate(data['w']):
            if j == wordIndex:
                continue
            inyearsimilarity.append((word,withinEachCosineSimilarities[i][wordIndex][j]))
        inyearsimilarity = sorted(inyearsimilarity, key=lambda key: key[1], reverse=True)
        similarities.append(inyearsimilarity[:10])
    return similarities
```

```python
def getJaccardIndex(similarWords):
    """
    each index corresponds to the jaccardindex from newest to that time
    """
    indices = []
    newest = set([x for x,y in similarWords[9]])
    for i in range(10):
        old = set([x for x,y in similarWords[i]])
        indices.append(100*len(newest.intersection(old))/len(newest.union(old)))
    return indices
```

```python
wordSimilarities = []
for index, word in enumerate(data['w']):
    wordSimilarities.append((word, getRelatedWords(index)))
```

```python
wordJaccardDistancesOverTime = []
for word, similarWords in wordSimilarities:
    wordJaccardDistancesOverTime.append((word, getJaccardIndex(similarWords)))
```

```python
# pearson correlations
# sortedSemanticChangeRate1
# semanticChangeCentury
# orderedDistances (eucledian)

cosDistance = [y for x, y in semanticChangeRate1]
timeDistance = [y for x,y in unsortedJaccard]
eucledianDistance = [y for x,y in presortedOrderedDist]

cosTimeCorr = pearsonr(cosDistance, timeDistance)
cosEucledianCorr = pearsonr(cosDistance, eucledianDistance)
TimeEucledianeCorr = pearsonr(eucledianDistance, timeDistance )
```
